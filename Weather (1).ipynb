```python
#top 10 hottest temperature values and their counts
top_temps=df['Temp_C'].value_counts().nlargest(10)
print(top_temps)
```

```python
#bar plot for top temperatures
plt.figure(figsize=(10,5))
sns.barplot(x=top_temps.index, y=top_temps.values, color="skyblue")
plt.xlabel("Temperature (C)")
plt.ylabel("Count")
plt.title("Top 10 hottest temperature values and their counts")
plt.xticks(rotation=45)
plt.show()
```

```python
#standard deviation of 'pressure' in this data
Pressure_std=df['Press_kPa'].std()
print(Pressure_std)
```

```python
#what is the variance of 'Relative humidity' in this data?
df['Rel Hum_%'].var()
```

```python
# Visualization for Relative Humidity Variance
plt.figure(figsize=(10, 5))
sns.boxplot(y=df["Rel Hum_%"], color="green")
plt.title("Boxplot of Relative Humidity")
plt.show()
```
